```python

import pandas as pd

cc = pd.read_csv('F:\projects\crx.data',header=None)
cc.columns=['Gender','Age','Debt','Married','BankCustomer','EducationLevel',
                                  'Ethnicity','YearsEmployed','Prior_Default','Employed','CreditScore','DriverLisence','Citizen'
                                 ,'ZipCode','Income','ApprovalStatus']
cc_apps=pd.DataFrame(cc)


cc_apps.head()
```

```python
for col in cc_apps.columns:
    if cc_apps[col].dtypes=='object' or 'int' or 'float':
        count=0
        count=[count + 1 for x in cc_apps[col] if x =='?']
        print (col +' ' + str(sum(count)))
```

```python
cc_apps[['Gender','Age','Married','BankCustomer','EducationLevel','Ethnicity','ZipCode']]=cc_apps[['Gender','Age','Married','BankCustomer','EducationLevel','Ethnicity','ZipCode']].replace('?',np.nan)
print(cc_apps.tail(17))
```

```python
cc_apps.fillna(cc_apps.mean(),inplace = True)

print(cc_apps.tail(17))
```

```python
print(cc_apps.isnull().sum())
```

```python

for col in cc_apps.columns:
  
    if cc_apps[col].dtypes=='object':
     
        cc_apps = cc_apps.fillna(cc_apps[col].value_counts().index[0])
       


print(cc_apps.isnull().sum())
print(cc_apps.tail(17))
```

```python

from sklearn import preprocessing
le=preprocessing.LabelEncoder()

for col in cc_apps.columns:
    
    if cc_apps[col].dtypes=='object':
        cc_apps[col]=le.fit_transform(cc_apps[col])
        
print(cc_apps.tail(17))        
```

```python

from sklearn.model_selection import train_test_split
import numpy as np
np.random.seed(3456)
cc_split = train_test_split(np.asmatrix(cc_apps), test_size = 227)
# print(cc_split)
cc_train_feature = cc_split[0][:, :13]
cc_train_label = np.ravel(cc_split[0][:, 13])
cc_test_feature = cc_split[1][:, :13]
cc_test_label = np.ravel(cc_split[1][:, 13])
print(cc_train_feature.shape)
print(cc_train_label.shape)
print(cc_test_feature.shape)
print(cc_test_label.shape)
```

```python

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
rescaledX_train = scaler.fit_transform(cc_train_feature)
rescaledX_test = scaler.fit_transform(cc_test_feature)
```

```python
from sklearn import linear_model
logreg=linear_model.LogisticRegression()
logreg.fit(rescaledX_train,cc_train_label)
```

```python
import sklearn.metrics as sklm

yscore=logreg.predict(rescaledX_test)

print("Accuracy of logistic regression classifier: ",logreg.score(rescaledX_test,cc_test_label))

conf=sklm.confusion_matrix(cc_test_label, yscore)
print('                  Confusion Matrix')
print('                 Score positive    Score negative')
print('Actual positive    %6d' % conf[0,0] + '             %5d' % conf[0,1])
print('Actual negative    %6d' % conf[1,0] + '             %5d' % conf[1,1])
print('')
```

```python
import sklearn.model_selection as ms
tol=[0.01, 0.001, 0.0001]
max_iter=[100,150,200]

param_grid=dict(tol=tol,max_iter=max_iter)

cc_split2 = train_test_split(np.asmatrix(cc_apps), test_size = 0)
# print(cc_split)
features = cc_split2[0][:, :13]
labels = np.ravel(cc_split2[0][:, 13])
print(features.shape)
print(labels.shape)
# cc_test_feature = cc_split[1][:, :13]

```

```python
grid_model=ms.GridSearchCV(estimator = logreg, param_grid = param_grid, cv = 5)
rescaledx=scaler.fit_transform(features)
grid_model_result = grid_model.fit(rescaledx,labels)

best_score, best_params = grid_model_result.best_score_ , grid_model_result.best_params_
print("Best: %f using %s" % (best_score,best_params))
```
